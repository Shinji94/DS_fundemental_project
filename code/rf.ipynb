```r
library(data.table)
#library(rattle)
library(rpart)
library(randomForest)
library(stringr) 
library(caret)
library(plyr)
library(dplyr)
library(stringr) 


#read dataset
setwd('/Users/xujialu/Desktop/460/')
train = fread('training_incidence.csv')
response = read.table('training_response.csv',header = F,sep = ',')

dim(train)
dim(response)
table(response)

#check missing datano missing data
sum(is.na(train)) 

try = train
try = as.data.frame(try)

#detect constant column and delete it  
lt = c()
n=1
for (i in 1:1392){
  if (length(levels(as.factor(try[,i])))<2){
    lt[n] = i 
    n = n+1
  }
  
}
lt
dim(try)
try = try[,-lt]
head(response)
#rename response as 'y'
DATA = cbind(response,try)
dim(DATA)
names(DATA)[DATA[,1]]="y"
```

```r

#delete duplicate observations
j = unique(DATA)
dim(j)

# randomForest to select features,use loop extend the forest to 5200 trees 
# randomForest (formula = as.factor(y) ~ .,
#              data = crs$dataset[crs$sample, c(crs$input, crs$target)],
#              ntree = 1000, mtry = 3)
for (i in 1:8){
library(randomForest)
rf1 <- randomForest(y ~ ., j, ntree=200+100*i,mtry = 3, norm.votes=true)

#save randomforest
save(rf1, file = "/Users/xujialu/Desktop/460/sb")
#load randomforest 
load("/Users/xujialu/Desktop/460/sb") 
summary(rf1)

    
    
k = rf1$importance
#this is the gini index
t =importance(rf1,type = 2)
head(importance(rf1,type=2))
    
#write the importance of each feature into a csv file
b=c("300_3","400_3","500_3","600_3","700_3","800_3","900_3","1000_3")
c=paste("/Users/xujialu/Desktop/460/",b[i],".csv")    
write.csv(t,c)

#error rate:which is not important
plot(rf1) 
#visualize feature importance 
varImpPlot(rf1)   
#try to predict using a certain randomforest
prediction <- predict(rf1,type="class") 
table(j$y,prediction)   

# one = filter(j,j$y == 1)
# dim(one)
table(j$y)
}


```
